```python
# The prompt and response columns are stored as JSON strings (lists of messages)
# You need to parse them into plain text

def parse_conversation(json_str):
    """Extract all text from a JSON-encoded list of messages."""
    try:
        messages = json.loads(json_str)
        if isinstance(messages, list):
            return " ".join(str(m) for m in messages)
        return str(messages)
    except:
        return str(json_str)

train['prompt_text'] = train['prompt'].apply(parse_conversation)
train['response_a_text'] = train['response_a'].apply(parse_conversation)
train['response_b_text'] = train['response_b'].apply(parse_conversation)

type(train['prompt_text'][0])
```
